```python
import turtle
import random
screen = turtle.Screen()
 
# Ellipse

def draw(rad):
     
    for i in range(2):
        turtle.circle(rad,90)
        turtle.circle(rad//2,90)
        

#Stars

def star_fil(step,angle):
    
    turtle.fillcolor("white")
    turtle.begin_fill()
    for i in range(5):
        turtle.forward(step)
        turtle.right(angle)
    turtle.end_fill() 
    
#Screen setup 

screen.setup(1500,1000)
screen.bgcolor('gray15')

for i in range(100):
    turtle.color('gray15')
    turtle.setx(random.randint(-725,725))
    turtle.sety(random.randint(-450,450))
    star_fil(10,144)

ind=500
val=180
turtle.color('gray15')

turtle.speed(0)
x,y=100,400


#Ellipse coordinates 

for i in range(0,150):
    turtle.setpos(x,y)
     
    turtle.seth(val)
    
    if(i<33):
        turtle.color("chocolate3")
    elif(i>=33 and i<67):
        turtle.color("chocolate2")
    elif(i>=67 and i<100):
        turtle.color("chocolate1")
    else:
        turtle.color("tan1")
     
    draw(ind)
    ind=ind-2
    x=x-1
    y=y-1
    
    
#Black Hole

turtle.setpos(-47.00,249.00)
turtle.fillcolor("black")
turtle.begin_fill()
for i in range(2):
    turtle.circle(220,90)
    turtle.circle(110,90)
turtle.end_fill()
```
